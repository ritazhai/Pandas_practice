### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url, sep=',')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
titanic.groupby('Sex').count()
data = titanic.groupby('Sex').count().PassengerId
data.plot(kind='pie',labels = ['Females', 'Males'], explode = (0.15 , 0), autopct = '%1.1f%%')
plt.axis('equal')
plt.title("Sex Proportion")
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)
lm.set(title = 'Fare vs Age')
```

### Step 7. How many people survived?

```python
titanic.Survived.sum()
```

### Step 8. Create a histogram with the Fare payed

```python
data2 = titanic.Fare
data2.plot.hist(bins=50)
plt.xlabel('Fare')
plt.title('Fare Histogram')
plt.show()
```

### BONUS: Create your own question and answer it.

```python
# survived percentage Females and males
Females_s = titanic[titanic.Sex == 'female'].Survived.sum()/titanic[titanic.Sex == 'female'].Survived.count()
Males_s = titanic[titanic.Sex == 'male'].Survived.sum()/titanic[titanic.Sex == 'male'].Survived.count()
proportions = [Females_s,Males_s]
plt.hist(proportions, range=(0,08))
plt.title("Survived Sex Proportion")
plt.show()
```
